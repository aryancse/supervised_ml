```python
import pandas as pd
```

```python
#loading data 
df=pd.read_csv('http://13.234.66.67/summer19/datasets/social.csv')
```

```python
#schema 
df.info()
```

```python
#top 5 rows 
df.head()
```

```python
#removing user id
filterdata=df.iloc[0:,2:].values
```

```python
#now we can apply training test
features=filterdata[0:,0:2]
```

```python
#label finding 
label=filterdata[0:,2]
```

```python
#training and testing
from sklearn.model_selection import train_test_split
```

```python
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```python
#now apply features scalling
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
#apply scaling in training data only
X=sc.fit_transform(X)
```

```python
X
```

```python
#testing data scaling 
x=sc.transform(x)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#random forest call with number of decision tree
Rclf=RandomForestClassifier(n_estimators=20)#by default it gives 10 trees
```

```python
#now train 
trained=Rclf.fit(X,Y)
```

```python
# now predicting 
output=trained.predict(x)
```
